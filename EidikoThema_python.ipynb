```python
#1st question

import pymongo
from pymongo import MongoClient
import os
#-----------------------------------------------------------------------
filepath = input("Give filepath for first question: ")
#filepath = "filepathtry.txt";
authors_ids=[];
AllAuthorsIdsof1doc=[];
f = open(filepath, "r")
for line in f:
    data = line;
    authors_ids.append(int(data));
#print(authors_ids)
f.close();
#-------------------------------------------------------------------------
# host variables for MongoDB
DOMAIN = 'localhost'
PORT = 27017

# create an instance of MongoClient()
client = MongoClient(
    host = DOMAIN + ":" + str(PORT)
)
# get the database_names from the MongoClient()
database_names = client.list_database_names()
#print ("\ndatabases:", database_names)
# create database & collection instances
db = client.MY_DB
collection = db.users
#--------------------------------------------------------------------------
collection.create_index("authors.id");
#---------------------------------------------------------------------------
cursor = collection.find({"authors.id": {"$in": authors_ids}})
results= {}
#print(db.collection.getShardDistribution())
for doc in cursor:
    #print(doc)
    key_Set=doc.keys()
    if ("references" in key_Set): 
        references = doc["references"]
        AllAuthorsof1doc = doc["authors"]
        AllAuthorsIdsof1doc=[]
        for a in AllAuthorsof1doc:
            AllAuthorsIdsof1doc.append(a["id"])
        for  ids in AllAuthorsIdsof1doc:
            if (ids in authors_ids):
                if ( ids in results.keys()):
                    s=[]
                    prevref=[]
                    prevref = results.get(ids);
                    s.extend(prevref);
                    s.extend(references);
                    d={ids: list( dict.fromkeys(s) )}
                    results.update(d);
                else:
                    d={ids: references}
                    results.update(d);
for one_ids in authors_ids :
    if( one_ids not in results.keys()) :
        d={one_ids: {}}
        results.update(d);
path = "1questionpython.txt";
if (os.path.exists(path)):
    f= open("1questionpython.txt","w+")
    f.truncate()
else :
    f= open("1questionpython.txt","w+")

    
for key in results.keys():
    f.write(str(key) )
    f.write(" ")
    f.write(str(len(results[key])))
    f.write(" ")
    for a in results[key]:
        f.write(str(a))
        f.write(" ")
    f.write("\n")
f.close()
```
